```python
import pandas as pd
import numpy as np
from sklearn import linear_model
from sklearn.preprocessing import LabelEncoder
```

```python
df = pd.read_csv("homeprice.csv")
df
```

```python
df.drop('locality_name',axis = 'columns')
df.drop('age',axis = 'columns')
```

```python
le_locality_name = LabelEncoder()
le_region_name = LabelEncoder()
le_construction_status = LabelEncoder()
le_house_type = LabelEncoder()
le_new_resale = LabelEncoder()
```

```python
df['locality_name_n'] = le_locality_name.fit_transform(df['locality_name'])
df['region_name_n'] = le_region_name.fit_transform(df['region_name'])
df['construction_status_n'] = le_construction_status.fit_transform(df['construction_status'])
df['house_type_n'] = le_house_type.fit_transform(df['house_type'])
df['new_resale_n'] = le_new_resale.fit_transform(df['new_resale'])
df
```

```python
ddf =df.drop(['locality_name','construction_status','region_name','house_type','new_resale','age'],axis='columns')
ddf
```

```python
reg = linear_model.LinearRegression()
reg.fit(dff[['value_per_sqft.values','area.values','total_rooms.values','total_beds.values','locality_name_n.values','region_name_n.values','construction_status_n.values','house_type_n.values','new_resale_n.values']],dff.price)
```

```python
reg.coef_
```

```python
reg.intercept_
```
